```python
from datetime import datetime, timedelta
# Function to calculate delay in seconds
def calculate_delay(expected_time, scheduled_time):
    # Convert time objects to datetime objects for the calculation
    expected_datetime = datetime.strptime(str(expected_time), '%H:%M:%S')
    scheduled_datetime = datetime.strptime(str(scheduled_time), '%H:%M:%S')
    # Adjust for the next day if expected time is past midnight relative to the scheduled time
    if expected_datetime < scheduled_datetime:
        expected_datetime += timedelta(days=1)
    # Calculate delay
    delay_seconds = (expected_datetime - scheduled_datetime).total_seconds()
    
    delay_minutes = delay_seconds / 60
    return delay_minutes
```

```python
# Apply the function to calculate delay for each row
df['Delay'] = df.apply(lambda row: calculate_delay(row['ExpectedTime_dt'], row['ScheduledTime_dt']), axis=1)
```
